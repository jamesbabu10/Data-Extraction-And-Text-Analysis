```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import openpyxl
wb=openpyxl.load_workbook("Output Data Structure.xlsx")
sh1=wb["Sheet1"]
```

```python
import requests 
import pandas as pd 
from bs4 import BeautifulSoup 
  
# link for extract html data 
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}
def getdata(url): 
    r = requests.get(url, headers=headers) 
    return r.text 
```

```python
syll='aeiou'
def count(word):
  count=0
  for i in word:
    if i in syll:
      count=count+1
  if word.endswith('e') or word.endswith('es') or word.endswith('ed'):
      count=count-1
  return(count)
```

```python
import pandas as pd
df = pd.read_csv('LoughranMcDonald_MasterDictionary_2020.csv', usecols = ['Word','Negative','Positive'])
positive_words=df[df.Positive!=0].Word.tolist()
negative_words=df[df.Negative!=0].Word.tolist()
```

```python
for row_num in range (2,171):
      htmldata = getdata(sh1.cell(row_num,2).value) 
      soup = BeautifulSoup(htmldata, 'html.parser') 
      data = '' 
      text=''
      for data in soup.find_all("h1"): 
        text=text+data.get_text()+" "
        #print(data.get_text()) 
      for data in soup.find_all("p"): 
        text=text+data.get_text()
        #print(data.get_text()) 
      text
      file = open(sh1.cell(row_num,2).value[8:].replace("/","_"), "w")
      file.write(text)  
      file.close()
      xyz=re.split(r'[.!?]+', text)
      number_of_sentences=len(xyz)
      #print(number_of_sentences)
      stop_words = set(stopwords.words('english'))
      up_ele = ["?",",",".","!",","]
      stop_words.update(up_ele)
      filtered_sentence=""
      words = text.split()
      for w in words :
          if w not in stop_words:
              filtered_sentence=filtered_sentence+' '+w
      #print(filtered_sentence)
      total_words=len(text.split())
      total_words_filtered=len(filtered_sentence.split())
      #print(total_words)
      words=word_tokenize(filtered_sentence) 
      positive_count=0
      negative_count=0
      for i in words:
        if i.upper() in positive_words:
          positive_count=positive_count+1
        elif i.upper() in negative_words:
          negative_count=negative_count+1
      #print(positive_count)
      #print(negative_count)
      complexword=0
      syll_count=0
      wordss=word_tokenize(text)
      for i in wordss:
        syll_count=syll_count+count(i)
        if count(i)>2:
          complexword=complexword+1
      #print(syll_count)
      #print(complexword)
      Personal_Pronouns_counts=0
      def counts(prono):
        return(sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(prono), text)))
      Personal_Pronouns="I i we We My my Ours ours us"
      for i in Personal_Pronouns.split(" "):
        Personal_Pronouns_counts=Personal_Pronouns_counts+counts(i)
      print("Positive Count:{}".format(positive_count))
      print("Negative Count:{}".format(negative_count))
      polarityScore=(positive_count-negative_count)/((positive_count+negative_count)+0.000001)
      print("Polarity Score:{}".format(polarityScore))
      subjectiveScore=(positive_count+negative_count)/((total_words)+0.000001)
      print("Subjective Score{}".format(subjectiveScore))
      Avg_sentance_lenght=total_words/number_of_sentences
      percentage_of_complex_words=complexword/total_words
      fog_index=0.4*(Avg_sentance_lenght+percentage_of_complex_words)
      print("Avg_sentance_lenght:{}".format(Avg_sentance_lenght))
      print("percentage_of_complex_words:{}".format(percentage_of_complex_words))
      print("fog_index:{}".format(fog_index))
      print("Avg_sentance_lenght:{}".format(Avg_sentance_lenght))
      print("complexword{}:".format(complexword))
      print("total_words_filtered{}:".format(total_words_filtered))
      syll_count_perWord=syll_count/total_words
      print("syll_count_perWord{}:".format(syll_count_perWord))
      print("Personal_Pronouns_counts{}:".format(Personal_Pronouns_counts))
      total_characters=len(text.replace(" ",""))
      avg_word_lenght=total_characters/total_words
      print("avg_word_lenght{}:".format(avg_word_lenght))
      i=row_num
      sh1.cell(i,3).value=positive_count
      sh1.cell(i,4).value=negative_count
      sh1.cell(i,5).value=polarityScore
      sh1.cell(i,6).value=subjectiveScore
      sh1.cell(i,7).value=Avg_sentance_lenght
      sh1.cell(i,8).value=percentage_of_complex_words
      sh1.cell(i,9).value=fog_index
      sh1.cell(i,10).value=Avg_sentance_lenght
      sh1.cell(i,11).value=complexword
      sh1.cell(i,12).value=total_words_filtered
      sh1.cell(i,13).value=syll_count_perWord
      sh1.cell(i,14).value=Personal_Pronouns_counts
      sh1.cell(i,15).value=avg_word_lenght
      print("url {} Done".format(row_num))
      print("---------------------------------------------------------------------")
wb.save("Output Data Structure.xlsx")
```
